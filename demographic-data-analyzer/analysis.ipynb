```python
# Read data from file
import pandas as pd
df = pd.read_csv("adult.data.csv")
df
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = df['race'].value_counts()
race_count
```

```python
# What is the average age of men?
average_age_men = round(df[df["sex"] == "Male"]["age"].mean(), 2)
print(average_age_men) 
```

```python
# What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = round(df[df['education'] == 'Bachelors'].shape[0] / df.shape[0] * 100, 2)
print(percentage_bachelors)
```

```python
#What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
adv_edu_high_salary = df[(df["education"].isin(["Bachelors", "Masters", "Doctorate"])) & (df["salary"] == ">50K")].shape[0]
adv_edu_high_salary_percentage = round(adv_edu_high_salary/df.shape[0]*100,2)
print(adv_edu_high_salary_percentage)

# What percentage of people without advanced education make more than 50K?
low_edu_high_salary =  df[(~df["education"].isin(["Bachelors", "Masters", "Doctorate"])) & (df["salary"] == ">50K")].shape[0]
low_edu_more_salary_percentage = round(low_edu_high_salary/df.shape[0]*100,2)
print(low_edu_more_salary_percentage)
```

```python
# with and without `Bachelors`, `Masters`, or `Doctorate`
adv_edu = ['Bachelors', 'Masters', 'Doctorate']
higher_education = df[df['education'].isin(adv_edu)]
lower_education = df[~df['education'].isin(adv_edu)]
```

```python
# percentage with salary >50K
higher_education_rich = round(higher_education[higher_education["salary"] == ">50K"].shape[0] / df.shape[0] * 100, 2)

lower_education_rich = round(lower_education[lower_education["salary"] == ">50K"].shape[0] / df.shape[0] * 100, 2)
lower_education_rich
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = df.groupby('hours-per-week').size().head(1).values[0]
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = (df[(df["salary"] == ">50K") & (df["hours-per-week"] == df["hours-per-week"].min())].shape[0]) / df.shape[0] * 100
num_min_workers
```

```python
rich_percentage = round(df[df["salary"] == ">50K"].shape[0] / df.shape[0] * 100, 2)
```

```python
# What country has the highest percentage of people that earn >50K?
highest_earning_country = df[df["salary"] == ">50K"].groupby("native-country").size().sort_values(ascending=False).head(1).index[0]
highest_earning_country_percentage = round(df[df["salary"] == ">50K"].groupby("native-country").size().sort_values(ascending=False).head(1).values[0] / df.shape[0] * 100, 2)
```

```python
# Identify the most popular occupation for those who earn >50K in India.
top_IN_occupation = df[df["native-country"] == "India"].groupby(["occupation", "salary"]).size().reset_index().sort_values(0, ascending=False).head(1)
```
